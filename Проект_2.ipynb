```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
data = [[13.5, 12.7, 12, 11.9, 11.5, 11.2, 10.8, 10.7, 10.6, 10.5],
        [251, 249, 248, 246, 242, 239, 235, 230, 228, 225],
        [0.91, 0.87, 0.85, 0.82, 0.79, 0.75, 0.7, 0.66, 0.62, 0.6],
        [2.54, 2.5, 2.45, 2.4, 2.37, 2.3, 2.27, 2.19, 2.05, 2],
        [6.3, 6.21, 6.15, 6, 5.8, 5.45, 5.05, 4.85, 4.5, 4.2],
        [18.2, 17.5, 17.1, 16.8, 16.1, 15.7, 15.2, 14.5, 14.3, 14],
        [134, 130, 128, 126, 122, 120, 117, 112, 108, 105],
        [64, 61, 58, 52, 49, 45, 40, 37, 34, 30],
        [4.25, 4.2, 4.18, 4.11, 4.05, 4, 3.91, 3.85, 3.77, 3.7],
        [1.8, 1.78, 1.7, 1.64, 1.59, 1.51, 1.45, 1.42, 1.4, 1.37]]
data = np.array(data)
X = np.arange(10).reshape(-1, 1)
model = LinearRegression()
trends = []
for i in range(10):
    y = data[:, i]
    model.fit(X, y)
    trend = model.coef_[0]
    trends.append(trend)
print("Тренды:", trends)
plt.plot(range(10), trends, marker='o')
plt.xlabel('Строка')
plt.ylabel('Тренд')
plt.title('Тренды матрицы данных')
plt.show()
```
