**Объединим данных и разделение на категориальные признаки**

```python

# Объединение данных
data = pd.concat([contract, internet, personal, phone], axis=1)
data.to_csv('combined_data.csv', index=False)
# Разделение на числовые и категориальные признаки
numeric_features = data.select_dtypes(include=['float', 'int'])
categorical_features = data.select_dtypes(include=['object'])


# Разделение на числовые и категориальные признаки
numeric_features = data.select_dtypes(include=['float', 'int'])
categorical_features = data.select_dtypes(include=['object'])
```

```python
# Создаем экземпляр класса OneHotEncoder
encoder = OneHotEncoder()

# Применяем One-Hot Encoding к данным
encoded_A = encoder.fit_transform(categorical_features).toarray()

# Выводим преобразованный датасет
print(encoded_A)
```

```python
scaler = StandardScaler()
scaled_d = scaler.fit_transform(numeric_features)
```

```python
preprocessed_data = pd.concat([pd.DataFrame(encoded_A), pd.DataFrame(scaled_d)], axis=1)
```

```python
target_variable = data['EndDate'].apply(lambda x: 1 if x == 'No' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(preprocessed_data, target_variable, test_size=0.2, random_state=42)
```

```python
# Создание и обучение модели логистической регрессии
logreg_model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
logreg_model.fit(X_train, y_train)

# Предсказание на тестовой выборке
y_pred = logreg_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
