```python
import matplotlib.pyplot as plt

from itertools import product

from sklearn.datasets import make_classification
from sklearn.datasets import make_blobs
from sklearn.datasets import make_gaussian_quantiles
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier

from sklearn.inspection import DecisionBoundaryDisplay

X, y = make_classification(
 n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
)
# Loading some example data

clf1 = KNeighborsClassifier(n_neighbors=1, weights = "distance", metric = "cityblock")
clf2 = KNeighborsClassifier(n_neighbors=2, weights = "uniform", metric = "cosine")
clf3 = KNeighborsClassifier(n_neighbors=3, weights = "distance", metric = "euclidean")
clf4 = KNeighborsClassifier(n_neighbors=4, weights = "uniform", metric = "haversine")
clf5 = KNeighborsClassifier(n_neighbors=5, weights = "distance", metric = "l1")
clf6 = KNeighborsClassifier(n_neighbors=6, weights = "uniform", metric = "l2")



clf1.fit(X, y)
clf2.fit(X, y)
clf3.fit(X, y)
clf4.fit(X, y)
clf5.fit(X, y)
clf6.fit(X, y)





# Plotting decision regions
f, axarr = plt.subplots(3, 2, sharex="col", sharey="row", figsize=(10, 8))
for idx, clf, tt in zip(
 product([0, 1, 2], [0, 1]),
 [clf1, clf2, clf3, clf4, clf5, clf6],
 ["KNN (k = 1), weights = distance, metric = cityblock",
 " KNN (k=2), weights = uniform, metric = cosine",
 "KNN (k=3), weights = distance, metric = euclidean",
 " KNN (k=4), weights = uniform, metric = haversine",
 "KNN (k=5), weights = distance, metric = l1",
 "KNN (k=6), weights = uniform, metric = l2"],
):
 DecisionBoundaryDisplay.from_estimator(
 clf, X, alpha=0.4, ax=axarr[idx[0], idx[1]], response_method="predict" )
 axarr[idx[0], idx[1]].scatter(X[:, 0], X[:, 1], c=y, s=20, edgecolor="k")
 axarr[idx[0], idx[1]].set_title(tt)

plt.show()
```
